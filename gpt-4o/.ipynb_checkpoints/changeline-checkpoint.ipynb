```python
import json, ast
import pandas as pd
from pathlib import Path
from openpyxl import load_workbook
from openpyxl.styles import Alignment
from openpyxl.utils import get_column_letter
```

```python
in_path  = "extracted.xlsx"
df       = pd.read_excel(in_path)
```

```python
json_cols = [c for c in df.select_dtypes("object").columns]

def prettify(js_str):
    """
    把 '{"a":"1","b":"2"}' → '{"a":"1",\n "b":"2"}'
    若解析失败，原样返回
    """
    if not isinstance(js_str, str): 
        return js_str
    try:
        # ast.literal_eval 支持单引号、None、True/False 等非标准 JSON
        obj = ast.literal_eval(js_str)
        # indent=0 会在每个元素后加 '\n', 分隔符置空去掉默认空格
        return json.dumps(obj, ensure_ascii=False, indent=0).replace('":', '":')  # 保留原冒号后无空格
    except Exception:
        return js_str  # 不是 JSON 时跳过

for col in json_cols:
    df[col] = df[col].apply(prettify)
```

```python
out_path = Path("Results_reformat.xlsx")
df.to_excel(out_path, index=False)
```

```python
wb = load_workbook(out_path)
ws = wb.active

for col in json_cols:                         # json_cols 是列名列表
    col_idx    = df.columns.get_loc(col) + 1  # 1-based index in Excel
    col_letter = get_column_letter(col_idx)   # 例如 1 → 'A', 27 → 'AA'

    # 给整列开启自动换行（跳过表头，从第 2 行开始）
    for cell in ws[col_letter][1:]:
        cell.alignment = Alignment(wrap_text=True)

wb.save(out_path)
print("√ 自动换行设置完成 →", out_path)
```
